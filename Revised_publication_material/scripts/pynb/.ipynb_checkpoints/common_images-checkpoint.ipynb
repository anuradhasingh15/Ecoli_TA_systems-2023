```python
#def function to apply threshold to toxin/antitoxions/and complete pairs
def apply_threshold(df):
    return df.apply(lambda x: [min(y, 1) for y in x])
```

```python
# Read in the data from CSV files and set the index to 'Strain'
t = pd.read_csv('/home/anuradha/Documents/anu/phd_project/TA-SYSTEMS/anaylsis-re/results-13-1-22/raw_files/hits.csv').set_index('Strain')
p = pd.read_csv('/home/anuradha/Documents/anu/phd_project/TA-SYSTEMS/anaylsis-re/results-13-1-22/raw_files/complete.csv').set_index('Strain')
at = pd.read_csv('/home/anuradha/Documents/anu/phd_project/TA-SYSTEMS/anaylsis-re/results-13-1-22/raw_files/partners.csv').set_index('Strain')
```

```python
# Apply the threshold to each dataframe
t_1 = apply_threshold(t)
p_1 = apply_threshold(p)
at_1 = apply_threshold(at)
```

```python
#function for adding ST column 
def add_ST_column(df):
    df_1 = df.copy()
    df_1['ST'] = df_1.index.str[2:-4].astype(int)
    first_column = df_1.pop('ST')
    df_1.insert(0, 'ST', first_column)
    return(df_1)
```

```python
#appply add ST column function to dataframes
t_1 = add_ST_column(t_1)
at_1 = add_ST_column(at_1)
p_1 = add_ST_column(p_1)
#at_1.head()
#t_1.head()
#p_1.head()
```

```python

def total_count(k):
    # Reset the index to default and set the multi-level index
    k_temp = k.reset_index().set_index(['ST', 'Strain'])
    
    # Compute the row-wise sum and rename the resulting column
    k_temp_sum = k_temp.sum(axis=1).to_frame().rename(columns={0: 'Total count'})
    
    # Reset the 'ST' level of the multi-level index and prepend 'ST' to the values
    k_temp_sum.reset_index(level='ST', inplace=True)
    k_temp_sum['ST'] = 'ST' + k_temp_sum['ST'].astype(str)
    
    # Reset the index to default and return the resulting dataframe
    return k_temp_sum.reset_index()
```

```python
t_count = total_count(t_1)
at_count = total_count(at_1)
p_count = total_count(p_1)
#t_count.head()
#at_count.head()
#p_count.head()
```

## <b> Boxplot, histogram, barplot and ecdf

```python
merged_count_df = pd.merge(pd.merge(t_count,at_count,on=['Strain', 'ST']),p_count,on=['Strain', 'ST'], indicator=True)
```

```python

#Histogram, ECDF plot and boxplot
fig = plt.figure(figsize=(10, 7), layout="constrained")
sns.set_style("white")

#two rows and two columns
gs = GridSpec(nrows=2, ncols=2, figure=fig)

#axis0 
ax0 = fig.add_subplot(gs[0, 0])
hue_colors = {"Complete":"crimson", "draft":"navy"}
sns.barplot(ax = ax0, data=set1_genomes_status, x="ST", y="Count", hue = 'Level',palette=hue_colors,
            hue_order=["Complete", "draft"], alpha=0.9, errorbar="sd")
ax0.set_title('A', fontsize = 10, fontweight="bold")#a) barplot depicting genome status
ax0.set_xlabel('Sequence types', fontsize = 10)
ax0.set_ylabel('No of isolates', fontsize = 10)
ax0.tick_params(axis='x',rotation=90, labelsize = 10)
ax0.tick_params(axis='y', labelsize = 10)
ax0.legend(bbox_to_anchor=(1.03, 1.05), fontsize = 8)
#despine all axes
sns.despine()

#axis1
ax1 = fig.add_subplot(gs[0, 1])
sns.histplot(ax=ax1, data=merged_count_df, x="Toxins", kde=True, color = 'crimson', alpha=0.6)
ax1.axvline(mean,color='green',label='Mean = 23.6') 
ax1.axvline(median,color='blue',label='Median = 23.0')
ax1.axvline(mode,color='black',label='Mode = 25')
ax1.set_title('B', fontsize = 10, fontweight="bold")#b) Histogram
ax1.set_xlabel('No of toxins', fontsize = 10)
ax1.set_ylabel('No of isolates', fontsize =10 )
ax1.tick_params(axis='both', labelsize = 10)
ax1.legend(bbox_to_anchor=(1.03, 1.05), fontsize = 8)
sns.despine()

#axis2 - ecdf plot
ax2 = fig.add_subplot(gs[1, 0])
#sns.ecdfplot(ax=ax1,data=a_sum, x="Total no of toxins", color = 'crimson')
sns.ecdfplot(ax=ax2,data = merged_count_df, x = "Toxins", color = 'crimson', stat = 'count', alpha=0.8)
sns.ecdfplot(ax=ax2,data=merged_count_df, x="Antitoxins", color = 'navy', stat = 'count', alpha=0.8)
sns.ecdfplot(ax=ax2,data=merged_count_df, x="T-A pairs", color = 'black', stat = 'count', alpha=0.8)
ax2.set_title('C', fontsize = 10, fontweight="bold")#c) ecdf plot
ax2.set_xlabel('Count', fontsize = 10 )
ax2.set_ylabel('Proportion', fontsize = 10)
ax2.tick_params(axis='both', labelsize = 10)
ax2.legend(["Toxins", "Antitoxins", 'T-A pairs'], bbox_to_anchor=(1, 0.5), fontsize = 8)
sns.despine()


#axis3 - boxplot stratified by STs
ax3 = fig.add_subplot(gs[1, 1])
sns.color_palette("tab10")
sns.boxplot(ax=ax3, data=merged_count_df, x = 'ST', y = "Toxins", notch=True, fliersize=2, whis=1.5)
ax3.set_title('D', fontsize = 10, fontweight="bold")#d) Total number of toxins per strain stratified by $Escherichia $ $coli$ STs
ax3.set_xlabel('Sequence types', fontsize = 10, )
ax3.set_ylabel('Toxins per isolate', fontsize = 10)
ax3.tick_params(axis='x',rotation=90, labelsize = 10)
ax3.tick_params(axis='y', labelsize = 10)
sns.despine()


#save plots
#plt.savefig('/home/anuradha/Documents/anu/phd_project/TA-SYSTEMS/TA_systems_clean/set1/pictures/publication/combined_fig1.tiff', dpi=300)
#plt.savefig('/home/anuradha/Documents/anu/phd_project/TA-SYSTEMS/TA_systems_clean/set1/pictures/pmrf_review/combined_fig1.tiff', dpi=300)
#plt.savefig('/home/anuradha/Documents/anu/phd_project/TA-SYSTEMS/TA_systems_clean/set1/pictures/combined_fig1.tiff', dpi=300)
plt.savefig('/home/anuradha/Documents/anu/phd_project/TA-SYSTEMS/TA_systems_clean/set1/Publication_material/pictures/main/Fig1.tiff', dpi=300)
plt.savefig('/home/anuradha/Documents/anu/phd_project/TA-SYSTEMS/TA_systems_clean/set1/Publication_material/pictures/allimages/Fig1.jpg', dpi=300)
```
