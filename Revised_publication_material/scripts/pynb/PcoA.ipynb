## <B> PCoA

```python
#Drop the strain information
t_d.drop(['Strain'],axis=1, inplace=True)
```

```python
#Split into X and Y datasets
X = t_d.drop(labels= "ST" , axis = 1)
y = t_d["ST"]
print(X.shape)
X.head()
```

```python
# Check for target balancing 
sns.set(font_scale=1)
plt.figure(figsize=(15,5))
sns.countplot(x='ST', data=t_d)
```

```python
classes=t_d['ST'].unique()
classes
```

```python
chi_score=chi2(X,y)
#chi_score
```

```python
p_value=pd.Series(chi_score[1],index=X.columns)
p_value.sort_values(ascending=True,inplace=True)
p_value
```

```python
len(p_value[p_value>0.05])
```

```python
# Distance Matrix
distance_matrix = squareform(pdist(X, metric="jaccard"))
print(distance_matrix)
```

```python
DM_dist = DistanceMatrix(distance_matrix, ids=X.index)
```

```python
print(X.index)
```

```python
PCoA = pcoa(DM_dist)
```

```python
permanova(DM_dist,y,permutations=999)
```

```python
# Calculate proportion of variance explained by each principal coordinate
eigenvalues = PCoA.eigvals
prop_var = eigenvalues / sum(eigenvalues)

# Create a data frame with eigenvalues and variance explained for top 10 principal coordinates
df = pd.DataFrame({"PC": np.arange(1, 11),
                   "ExplainedVarianceRatio": prop_var[:10]})


# Create a list of PC labels
pc_labels = ["PC{}".format(i) for i in range(1, 11)]

# Plot the top 10 principal coordinates and their explained variance ratio
fig, ax = plt.subplots()
plt.plot(df["PC"], df["ExplainedVarianceRatio"], 'o-', linewidth=2, color='blue')
#ax.bar(df["PC"], df["ExplainedVarianceRatio"])
ax.set_xticks(np.arange(1, 11, 1))
ax.set_xticklabels(pc_labels)
ax.set_xlabel("Principal coordinates")
ax.set_ylabel("Explained variance ratio")
#plt.savefig("/home/anuradha/Documents/anu/phd_project/TA-SYSTEMS/TA_systems_clean/set1/pictures/scree_plot.tiff", dpi = 300)
#plt.savefig("/home/anuradha/Documents/anu/phd_project/TA-SYSTEMS/TA_systems_clean/set1/pictures/publication/scree_plot.tiff", dpi = 300)
#plt.savefig("/home/anuradha/Documents/anu/phd_project/TA-SYSTEMS/TA_systems_clean/set1/pictures/pmrf_review/scree_plot.tiff", dpi = 300)
plt.savefig('/home/anuradha/Documents/anu/phd_project/TA-SYSTEMS/TA_systems_clean/set1/Publication_material/pictures/supplemantary/SFig1.tiff', dpi=300)
plt.savefig('/home/anuradha/Documents/anu/phd_project/TA-SYSTEMS/TA_systems_clean/set1/Publication_material/pictures/allimages/SFig1.tiff', dpi=300)
```
