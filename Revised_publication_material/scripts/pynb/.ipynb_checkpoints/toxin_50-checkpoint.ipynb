```python
def t_percentage(df,tag):

    df_h=df.copy()
    #df_h.head()
    x=list(df_h.columns)
    #print(x)
    size=df_h.shape[0]
    #print(size)
    df_toxin=pd.DataFrame(columns=['Toxins',tag])
    #print(df_toxin.head())
    for i in x:
        #print(df_h[i])
        s=df_h[i].astype(int).sum()
        #print(s)
        p=(s/size)*100

        df_toxin = df_toxin.append({'Toxins': i,tag:p}, ignore_index=True)



    return(df_toxin)
```

```python
#Calculate percentage of gene presence in each ST
toxin_list=[]
for i in df_list:
    d=globals()[i].copy()
    d.drop('ST', axis=1,inplace=True)
    x=i.replace('df','t')
    y=i.replace('_df','')
    toxin_list.append(x)
    globals()[x]=t_percentage(d,y)
    print(x,globals()[x].shape)
```

```python
def merge_dataframe(list):
    l=list.copy()
    x=l.pop(0)
    x_df=globals()[x]
    for i in l:
        x_df=pd.merge(x_df,globals()[i],how='outer',on='Toxins').fillna(0)
    return(x_df)
```

```python
# Merge the dataframes
t_df=merge_dataframe(toxin_list)
t_df
```

```python
def host_classification(k):
    k = k.replace(['None', 'missing', 'not applicable','not available',
                           'Unknown', 'not collected', 'Not Available', 'Not applicable'], 'UNKW')
    k = k.replace(['Human', 'Homosapiens', 'Homo sapiens; 55 year old', 'Homo sapiens'], 'Homo_sapiens')
    k = k.replace(['Bos taurus', 'cattle', 'bovine', 'cow', 'Dairy cow'], 'Cattle')   
    k = k.replace(['Meleagris gallopavo', 'Turkey'], 'turkey')
    k = k.replace(['Sus scrofa', 'Sus scrofa scrofa'], 'wild_Boar')
    k = k.replace(['Sus scrofa domesticus', 'pig','swine', 'Swine', 'Pig'], 'pig')
    k = k.replace(['Gallus gallus', 'chicken', 'Gallus gallus domesticus', 'broiler chicken',
                   'egg laying hen', 'Gallus gallus domesticus (chicken)'], 'Chicken')
    k = k.replace(['dog', 'Canis lupus familiaris', 'canine', 'pet dog', 'Dog'], 'Dog')
    k = k.replace(['Ovis aries'], 'Sheep')
    k = k.replace(['Giant Panda', 'Deer', 'Odocoileus virginianus','Macaca nigra (Celebes ape)', 'swallow',
                   'Larus sp. (gull)', 'Avian','Daphnia obtusa','pheasant','Duck', 'Bivalve mollusk','rat', 
                   'Spheniscus magellanicus', 'soldier','wild bird',
                   'animal','Larus michahellis','Bos mutus','Nanger dama','Larus sp.',
                   'cat', 'Ailuropoda melanoleuca','mink', 'fly',  'calamain deer','duck', 'Sewage','Threskiornis spinicollis'
                  , 'demoiselle crane', 'mink'],
                  'Others')

        

    return(k)
```
